```python
#도전
midwest = pd.read_csv('midwest.csv')
```

```python
midwest.head()
```

```python
midwest =  midwest.rename(columns ={'poptotal':'total', 'popasian':'asian'})
midwest.head()
```

```python
midwest['ratio'] = midwest['asian'] / midwest['total'] * 100
midwest['ratio'].plot.hist()
```

```python
mean = midwest['ratio'].mean()
```

```python
mean
```

```python
midwest['group'] = np.where(midwest['ratio'] > mean, 'large', 'small')
```

```python
count_group = midwest['group'].value_counts()
count_group
```

```python
count_group.plot.bar(rot=0)
```
