```python
moon = open('speech_moon.txt', encoding= 'UTF-8').read()
moon
```

```python
import re
moon = re.sub('[^가-힣]', ' ', moon)
moon
```

```python
import konlpy
```

```python
hannanum = konlpy.tag.Hannanum()

hannanum.nouns('대한민국의 영토는 한반도와 그 부속도서로 한다')
```

```python
nouns = hannanum.nouns(moon)
nouns
```

```python
import pandas as pd
df_word = pd.DataFrame({'word':nouns})
df_word
```

```python
df_word['count'] = df_word['word'].str.len()
df_word
```

```python
df_word = df_word.query('count >= 2')
df_word.sort_values('count')
```

```python
df_word = df_word.groupby('word', as_index= False)\
                 .agg(n = ('word','count'))\
                 .sort_values('n', ascending = False)
df_word
```

```python
top20 = df_word.head(20)
top20
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.rcParams.update({'font.family': 'Malgun Gothic',
                     'figure.dpi':'120',
                     'figure.figsize':[6.5, 6]})

sns.barplot(data = top20, y = 'word', x = 'n')
```

```python
font = 'DoHyeon-Regular.ttf'
```

```python
dic_word = df_word.set_index('word').to_dict()['n']
dic_word
```

```python
from wordcloud import WordCloud
wc = WordCloud(random_state=1234,
              font_path = font,
              width = 400,
              height = 400,
              background_color= 'white')
```

```python
img_wordcloud = wc.generate_from_frequencies(dic_word);

plt.figure(figsize = (10, 10))
plt.axis('off')
plt.imshow(img_wordcloud)
```

```python
import PIL
icon = PIL.Image.open('twit.png')
```

```python
import numpy as np
img = PIL.Image.new('RGB', icon.size, (255, 255, 255))
img.paste(icon, icon)
img = np.array(img)
```

```python
wc = WordCloud(random_state = 1234,
               font_path = font,
               width = 400,
               height = 400,
               background_color = 'white',
               mask = img)
img_wordcloud = wc.generate_from_frequencies(dic_word)

plt.figure(figsize = (10, 10))
plt.axis('off')
plt.imshow(img_wordcloud);
```

```python
df = pd.read_csv('news_comment_BTS.csv', encoding='utf-8')
```

```python
df.info()
```

```python
df['reply'] = df['reply'].str.replace('[^가-힣]', ' ', regex = True)
df['reply'].head()
```

```python
kkma = konlpy.tag.Kkma()
```

```python
nouns = df['reply'].apply(kkma.nouns)
```

```python
nouns
```

```python
nouns = nouns.explode()
nouns
```

```python
df_word = pd.DataFrame({'word':nouns})
df_word
```

```python
df_word['count'] = df_word['word'].str.len()
df_word = df_word.query('count >= 2')
df_word
```

```python
df_word = df_word.groupby('word', as_index= False)\
                 .agg(n = ('word', 'count')).sort_values('n', ascending= False)
df_word
```

```python
top20 = df_word.head(20)
top20
```

```python
plt.rcParams.update({'figure.figsize': [6.5, 6]})
sns.barplot(data = top20, y='word', x = 'n')
```
